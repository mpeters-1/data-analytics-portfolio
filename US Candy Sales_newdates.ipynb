```python
import pandas as pd
```

```python
import numpy as np
```

```python
# Loading dataset
df = pd.read_csv("Candy_Sales.csv")
```

```python
# Converting shipping dates
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
# Randomizing shipment dates
def generate_ship_date(order_date):
    # 2 to 5 days after order placed
    if np.random.rand() < 0.8:
        days_delayed = np.random.randint(2, 6)
    else:
    # 7 to 15 days after order placed
        days_delayed = np.random.randint(7, 16)
    return order_date + pd.Timedelta(days=days_delayed)
```

```python
# New column
df["New Ship Date"] = df["Order Date"].apply(generate_ship_date)

print(df[["Order Date", "Ship Date", "New Ship Date"]].head(10))
```

```python
# New csv
df.to_csv("Candy_Sales_NewShipDates.csv", index=False)

print("File saved as Candy_Sales_NewShipDates.csv")
```

```python
# Making a copy
df_export = df.copy()

# Convert to YYYY-MM-DD string format
df_export["Order Date"] = df_export["Order Date"].dt.strftime("%Y-%m-%d")
df_export["Ship Date"] = df_export["Ship Date"].dt.strftime("%Y-%m-%d")
df_export["New Ship Date"] = df_export["New Ship Date"].dt.strftime("%Y-%m-%d")

# Exporting to csv
df_export.to_csv("Candy_Sales_NewShipDates.csv", index=False)

print("File saved with dates in YYYY-MM-DD format")
```

```python
# Reorder columns so New Ship Date is right after Ship Date
cols = list(df.columns)
ship_idx = cols.index("Ship Date")
cols.remove("New Ship Date")
cols.insert(ship_idx + 1, "New Ship Date")
df = df[cols]

# (Optional) format dates as YYYY-MM-DD so Excel doesn’t mess them up
df_export = df.copy()
for col in ["Order Date", "Ship Date", "New Ship Date"]:
    df_export[col] = df_export[col].dt.strftime("%Y-%m-%d")

# Save to CSV with new column order
df_export.to_csv("Candy_Sales_With_New_ShipDates.csv", index=False)

print("CSV saved with 'New Ship Date' right after 'Ship Date'")
```
